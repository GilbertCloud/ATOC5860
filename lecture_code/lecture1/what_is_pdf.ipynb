```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd  # library for data analysis for text files (everything but netcdf files)
%matplotlib inline 
```

```python
#===============================================
# draw random values from normal distribution
#===============================================
x = np.random.normal(0,1,size = (5,))   ## small sample
y = np.random.normal(0,1,size = (3000,)) ## big sample
```

```python
plt.xlabel('value')
plt.ylabel('arbitrary')
plt.yticks([])
plt.title('values drawn from a normal distribution')
plt.xlim(-3,3)
plt.plot(x,np.arange(0,np.size(x)),'.r', markersize = 5)
plt.plot(y,np.arange(0,np.size(y)),'.k', markersize = 5)
xinc = np.arange(-4.5,5.5,.5)
for ind, val in enumerate(xinc[1:]):
    plt.plot([val - 0.25, val-0.25], [0, np.size(y)],'-', color = 'gray')
plt.show()
```

```python
## Plot Histogram Probability Distribution Function
hx = np.histogram(x,xinc)
hy = np.histogram(y,xinc)
plt.xlabel('Value')
plt.ylabel('Count')
plt.bar(hx[1][:-1],hx[0],edgecolor = 'r', color = [], width = .4, linewidth = 2)
plt.bar(hy[1][:-1],hy[0],edgecolor = 'k', color = [], width = .4, linewidth = 2)
plt.show()
```

```python
## Plot Histogram Probability Distribution Function
## Question for you -- How does the plot below differ from the one above?  Check out the x-axis in particular!

plt.rcParams["figure.figsize"] = [4,4];
plt.hist(y, bins=xinc, histtype='bar', rwidth=0.5, alpha=1,color='dodgerblue');
plt.hist(x, bins=xinc, histtype='bar', rwidth=0.5, alpha=1,color='firebrick');
plt.xlabel('Value');
plt.ylabel('Count');
#plt.xlim([0,4]);
#plt.ylim([0,1]);
plt.legend(['small', 'big'],loc='upper right',frameon=True,markerscale=0.5,framealpha=1);
```

```python
## Plot Probability Density Function

density_option=True
## If Density = True, draw and return a probability density: each bin will display the bin's 
## raw count divided by the total number of counts and the bin width 
##(density = counts / (sum(counts) * np.diff(bins))), 
## so that the area under the histogram integrates to 1 (np.sum(density * np.diff(bins)) == 1).

plt.rcParams["figure.figsize"] = [4,4];
plt.hist(y, bins=xinc, histtype='bar', rwidth=0.5, alpha=1,color='dodgerblue',density=density_option);
plt.hist(x, bins=xinc, histtype='bar', rwidth=0.5, alpha=1,color='firebrick',density=density_option);
plt.xlabel('Value');
plt.ylabel('Probability Density');
plt.legend(['small', 'big'],loc='upper right',frameon=True,markerscale=0.5,framealpha=1);
```

```python
## Plot Cumulative Distribution Function
cum_option=True

plt.rcParams["figure.figsize"] = [4,4];
plt.hist(y, bins=xinc, histtype='bar', rwidth=0.5, alpha=1,color='dodgerblue',density=density_option,cumulative=cum_option);
plt.hist(x, bins=xinc, histtype='bar', rwidth=0.5, alpha=1,color='firebrick',density=density_option,cumulative=cum_option);
plt.xlabel('Value');
plt.ylabel('Probability Density');
plt.legend(['small', 'big'],loc='lower right',frameon=True,markerscale=0.5,framealpha=1);
```
