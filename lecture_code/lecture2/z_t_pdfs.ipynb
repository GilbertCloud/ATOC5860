```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import scipy.io as sio
import pandas as pd  # library for data analysis for text files (everything but netcdf files)
%matplotlib inline
```

```python
### Compare Z versus t distributions.  
# Increase or decrease the degrees of freedom (N-1) to see how it affects the distribution shapes
# How large does N have to be for the distributions to look the same?

df = 6  ## degrees of freedom (N-1) - CHANGE ME
x = np.arange(-6,6,.01)
z = stats.norm.pdf(x,0,1)
t = stats.t.pdf(x,df)

f=plt.figure(figsize=(20,6))
plt.plot(x,z, color = 'red', label = 'Z');
plt.plot(x,t,linestyle = '-', color = 'black', label = r"t ($\nu$ = "+str(df)+')');

plt.title('Z and Student-t probability density functions',fontsize=18)
plt.ylabel('Probability of Occurrence: f(Z) or f(t)',fontsize=18)
plt.legend(frameon = 0,fontsize=18)

plt.xlim(-5,5);
plt.ylim(0,.45);
plt.yticks(np.arange(0,.5,.1));
```
